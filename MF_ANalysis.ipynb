```python
import pandas as pd
import numpy
import os
import matplotlib.pyplot as plt
%matplotlib inline
import time
```

```python
start = time.time()
main_df = pd.DataFrame()
for root, _, files in os.walk('./data'):
    for file in files:
        file_path = os.path.join(root, file)
        if (file_path.endswith('.csv')):
            df = pd.read_csv(file_path)
            main_df = pd.concat([main_df, df], ignore_index=True, axis=0, sort=True)
print(start)
print(time.time() - start)
```
